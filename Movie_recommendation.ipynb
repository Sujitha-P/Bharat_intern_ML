```python
import pandas as pd
```

```python
rating = pd.read_csv('/content/ratings_data.csv')
movie = pd.read_csv('/content/movies_data.csv')
data = pd.merge(movie,rating).drop(['timestamp'],axis=1)
print (data.shape)
data.head()
```

```python
user_ratings = data.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0,axis=1)
print (user_ratings.shape)
user_ratings.head()
```

```python
matrix = user_ratings.corr(method='pearson')
matrix.head()
```

```python
def recommend(movie_name, user_rating):
  similar = matrix[movie_name]*(user_rating-2.5)
  similar = similar.sort_values(ascending=False)
  return similar
similar_movies = pd.DataFrame()
movie_name = str(input())
user_rating = float(input())
similar_movies = similar_movies.append(recommend(movie_name, user_rating))
similar_movies.head()
similar_movies.sum().sort_values(ascending=False).head(10)
```
